```python
import os
import sys
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re

def extractTable(city_url):
    url = city_url
    page = requests.get(url)
    soup = BeautifulSoup(page.content,"html.parser")
    city = soup.find("div",class_="titulo")
    city=city.text.split("\n")[1]

    mt5=soup.find("p",class_="mt5")
    latitude = re.findall("\d+\.\d+",mt5.text)[0]
    longitude = re.findall("\d+\.\d+",mt5.text)[1]

    data = []

    table= soup.find('table', attrs={'class':'medias'})

    rows = table.find_all('tr')

    for row in  rows:
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        data.append(cols)
    
    df = pd.DataFrame(data, columns = ['Year','T','TM','Tm','PP','V','RA','SN','TS','FG','TN','GR'])

    cities = []
    for i in range (df.shape[0]):
                    cities.append(city)

    latitudes = []
    for i in range (df.shape[0]):
                    latitudes.append(latitude)

    longitudes = []
    for i in range (df.shape[0]):
                    longitudes.append(longitude)

    x= list(zip(cities,latitudes, longitudes))

    df_1 = pd.DataFrame(x, columns=['City','Latitude','Longitude'])

    completed_df = pd.concat([df_1,df],axis= 1)

    return completed_df

cities = pd.read_excel(r'C:\Users\abhij\python_project\url.xlsx')

final = pd.DataFrame()
for index, row in cities.iterrows():
    temp = extractTable(row['Url'])
    final = pd.concat([final,temp], axis=0)
print(final)
```
